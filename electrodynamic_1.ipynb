```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import time
import datetime
%matplotlib inline
```

# 画电多极矩的图

```python
#创建二维正方形空间点（x,y）
x1=np.linspace(-2,2,500)
x = np.zeros(shape=(500,500))
for i in range(500):
    x[i,:] = x1
y = np.zeros(shape=(500,500))
for i in range(500):
    y[:,i] = x1
```

```python
#作圆 半径为1圆心在原点
r = 1.0
a, b = (0., 0.)
theta = np.arange(0, 2*np.pi, 0.01)
x_c = a + r * np.cos(theta)
y_c = b + r * np.sin(theta)

```

```python
#球外势公式
def fai_1(x,y,e,e0,E0,r):
    fai1 = -E0*x+( (e-e0)*x*E0*r**2 / \
                  ((e+2*e0)*(x**2+y**2)**(3/2)) )
    return fai1
```

```python
#参数
e0 = 8.854187817*10**(-12) #介电常数
e = (1.3)*e0 #10**4
E0 = 1 #外场

fai1 = fai_1(x,y,e,e0,E0,r)#球外势
fai2 = -(3*e0)*x*E0/(e+2*e0) #球内势

```

```python

Phi_final = fai1
fai1_2=[]
x_2=[]
y_2=[]
for i in range(500):
    for j in range(500):
        if x[i,j]**2 + y[i,j]**2 < 1:#将球内空间点的势替换成fai2
            Phi_final[i,j] = fai2[i,j]
        else: #专门画球外势的二极矩部分
            x_2.append(x[i,j])
            y_2.append(y[i,j])
            fai12=( (e-e0)*x[i,j]*E0*r**2 / \
                  ((e+2*e0)*(x[i,j]**2+y[i,j]**2)**(3/2)) )
            fai1_2.append(fai12)
x_2=np.array(x_2)
y_2=np.array(y_2)
fai1_2=np.array(fai1_2)
```

```python
#作图
plt.figure(figsize=(12.5,10))
#plt.figure()
plt.xlabel('x',fontsize=12)
plt.ylabel('y',fontsize=12)

cm = plt.cm.get_cmap('pink')

cirlce = plt.plot(x_c,y_c,'w')
sc = plt.scatter(x,y,c = Phi_final, s=7,edgecolors = 'None',cmap=cm)
clb = plt.colorbar(sc)
clb.ax.set_title('$phi$',fontsize=12)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)


# plt.scatter(starcatalog_fram[teff],starcatalog_fram[err_teff],\
#             c='red',s=5, edgecolors='None', alpha=0.5)

plt.legend()
plt.savefig('1p3ele_1005.png',dpi=200)
plt.show()
```
