```python
#read AB
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
import math
path = '/Users/SY/Desktop/obic/data/H1991.fit'
H = fits.open(path)
```

```python
H[1].header
```

```python
f = H[1].data
```

```python
spt = f['SpType']
spt[1]
```

```python
Bs = []
Num = []

for i in range(len(spt)):
    if len(spt[i]) != 0:
        if spt[i][2] == 'B':
            if spt[i][3].isdigit(): 
                Num.append(i)
                Bs.append(spt[i])
```

```python
Num
```

```python
print(set(Bs))
```

```python
def judge(ele):
    e_figure = 0
    if '.' in ele:
        e_figure = ele[3:6]
        if '+' in e_figure:
            e_figure = e_figure[0]
    
        if 'IV' in e_figure:
            e_figure = e_figure[0]
    else:
        if ele != '  BepShell' and ele != 'sdBe+gM3+Q':
            e_figure = ele[3]
    if e_figure != 0:
        return e_figure
    else:
        return -1
```

```python

DE = []
RA = []
SPT = []
SPT_pure = []
for i in Num:
    j = judge(spt[i])
    
    if j != -1:
        DE.append(f['_DEJ2000'][i])
        RA.append(f['_RAJ2000'][i])
        SPT.append(f['SpType'][i])
        SPT_pure.append(j)
out_f = np.vstack((np.array(DE),np.array(RA),np.array(SPT),np.array(SPT_pure))).T
```
