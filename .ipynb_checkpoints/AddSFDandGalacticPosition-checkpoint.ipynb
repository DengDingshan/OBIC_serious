```python
# Only GP
l = []; b = [];
for j in range(len(df)):
    c1 = coord.ICRS(ra = df['RA'].values[j]*u.degree, dec = df['DEC'].values[j]*u.degree)
    gc1 = c1.transform_to(coord.Galactic)
    l.append(gc1.l.value)
    b.append(gc1.b.value)
```
