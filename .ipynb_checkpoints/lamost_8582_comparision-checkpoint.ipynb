# 数据的引入和format (Alternative)

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import time
import datetime
from scipy.optimize import curve_fit
%matplotlib inline

data_dir = '/Users/SY/Desktop/obic/data/'
```

```python
save = data_dir + 'pictures/'

i = datetime.datetime.now()
date_now = "%s_%s_%s_" % (i.year, i.month, i.day) 

savename1 = save + date_now  + 'teff_tlusty'+'.png'
savename2 = save + date_now  + 'teff_val'+'.png'
savename3 = save + date_now  + 'logg_tlusty'+'.png'
```

```python
starcatalog_fram = pd.read_csv(data_dir + 'lamost_8582.csv')

starcatalog_fram_fil = starcatalog_fram.fillna(0)

print(starcatalog_fram.columns)
```

```python
teff_1 = starcatalog_fram['TEFF'].values
teff_2 = starcatalog_fram['teff_val'].values
teff_3 = starcatalog_fram['Teff_tlusty'].values

logg_1 =starcatalog_fram['LOGG'].values 
logg_2 =starcatalog_fram['logg_tlusty'].values 
#starcatalog_fram
```

```python
z1 = np.polyfit(teff_1, teff_3, 1)  #一次多项式拟合，相当于线性拟合
p1 = np.poly1d(z1)
print(z1,p1)



teffmax=np.max([np.max(teff_1),np.max(teff_3)])


x=np.linspace(0,teffmax,10000)
y1=z1[0]*x+z1[1]
y2=x


plt.figure(1)
plt.xlabel('$T_\mathrm{eff}(K)$')
plt.ylabel('$T_\mathrm{eff}(tlusty)(K)$')
plt.scatter(teff_1,teff_3,s=1,c='k',alpha=0.5)
plt.plot(x,y2,'r',label='y=x')
plt.plot(x,y1,label='fitting curve')
plt.xlim(0,teffmax)
plt.ylim(0,teffmax)

plt.legend(loc='lower right')
plt.savefig(savename1,dpi=200)
plt.show()
```

```python
z1 = np.polyfit(teff_1, teff_2, 1)  #一次多项式拟合，相当于线性拟合
p1 = np.poly1d(z1)
print(z1,p1)



teffmax=np.max([np.max(teff_1),np.max(teff_2)])


x=np.linspace(0,teffmax,10000)
y1=z1[0]*x+z1[1]
y2=x


plt.figure(1)
plt.xlabel('$T_\mathrm{eff}(K)$')
plt.ylabel('$T_\mathrm{eff}(val)(K)$')
plt.scatter(teff_1,teff_2,s=1,c='k',alpha=0.5)
plt.plot(x,y2,'r',label='y=x')
plt.plot(x,y1,label='fitting curve')
plt.xlim(0,teffmax)
plt.ylim(0,teffmax)

plt.legend(loc='upper left')
plt.savefig(savename2,dpi=200)
plt.show()
```

```python
z1 = np.polyfit(logg_1, logg_2, 1)  #一次多项式拟合，相当于线性拟合
p1 = np.poly1d(z1)
print(z1,p1)



loggmax=np.max([np.max(logg_1),np.max(logg_2)])


x=np.linspace(0,loggmax,10000)
y1=z1[0]*x+z1[1]
y2=x


plt.figure(1)
plt.xlabel('$logg$')
plt.ylabel('$logg(tlusty)$')
plt.scatter(logg_1,logg_2,s=1,c='k',alpha=0.5)
plt.plot(x,y2,'r',label='y=x')
plt.plot(x,y1,label='fitting curve')
plt.xlim(2,loggmax)
plt.ylim(2,loggmax)

plt.legend(loc='lower right')
plt.savefig(savename3,dpi=200)
plt.show()
```
