```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import time
%matplotlib inline

data_dir = '/Users/SY/Desktop/obic/data/'
```

```python
#read as pandas

gosss_fram = pd.read_csv(data_dir + 'gosss_all.csv')
judge_fram = pd.read_csv(data_dir + 'spt-teff.csv')
#starcatlog_fram = starcatlog_fram.ix[:,:517]
judge_fram.columns=['spt','s','Ia+','Ia','Iab','Ib','II','III','IV','V']
print(gosss_fram)
```

```python
#bu qi data
gosss_fram_fil = gosss_fram.fillna(0)
```

```python
spt = gosss_fram_fil['STv3']
new_spt = list()
#modify spt
for s in spt:
    #print(type(s))
    if 'N' in s or 'C' in s:
        new_s = s[0] + s[2:]
    else:
        new_s = s[0:]
    if ':' in new_s:
        new_s = new_s[:-1]
    if '-' in new_s:
        new_s = new_s[:-2]
    if '9.2' in new_s:
        new_s = new_s[0] + '9'+ new_s[4:]
    #print(type(new_s))
    new_spt.append(new_s)
    
new_spt = np.array(new_spt)
gosss_fram_fil['STv3'] = new_spt
#print(type(gosss_fram_fil['STv3'].values))

print(gosss_fram_fil)
```

```python
#modify type
gosss_fram_fil = gosss_fram_fil[(gosss_fram_fil['LCv3']!=0)]
tp = gosss_fram_fil['LCv3']
new_tp = list()
#modify spt
for t in tp:
    new_t = t
    if ':' in t:
        new_t = t[:-1]
    if '-' in new_t:
        new_t = new_t[0:1]
    #print(type(new_s))
    new_tp.append(new_t)
    
new_tp = np.array(new_tp)
gosss_fram_fil['LCv3'] = new_tp
```

```python
gosss_fram_fil = gosss_fram_fil[((gosss_fram_fil['STv3']!='O2')\
                                & (gosss_fram_fil['STv3']!='O2.5')\
                                & (gosss_fram_fil['STv3']!='O3')\
                                & (gosss_fram_fil['STv3']!='O3.5')\
                                & (gosss_fram_fil['LCv3']!='I'))]
```

```python
# divert spt to teff
#teff=pd.DataFrame(columns=['teff'])
teff = list()
special = gosss_fram_fil['LCv3']
for n in range(len(gosss_fram_fil['STv3'])):
    i,k = gosss_fram_fil['STv3'].values[n],gosss_fram_fil['LCv3'].values[n]
    #print(i,k)
    for m in range(len(judge_fram['spt'])):
        j = judge_fram['spt'].values[m]
        #print(j)
        if i == j:
            for l in judge_fram.columns:
                #print(l)
                if k == l:
                    #print(judge_fram[l].values[m])
                    teff.append(judge_fram[l].values[m])

teff = np.array(teff)
gosss_fram_fil['TEFF'] = teff
gosss_fram_fil['ERR_TEFF'] = 0.000000001
                        
        
```

```python
#change columns'name
gosss_fram_fil.rename(columns={'RA':'ra', 'RAd':'RA', 'decd':'DEC'}, inplace = True)
```

```python
gosss_fram_fil.to_csv(data_dir + 'gosss_new.csv',columns = ['TEFF','ERR_TEFF','RA','DEC'])
```
